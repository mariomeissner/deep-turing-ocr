```
img_h = 30
img_w = 170
num_h = 10
num_w = 10
```

```
white = Image.new('L', (img_w, img_h), color=255)
# display(white)
```

```
# for i in range(len(x_train[:50])):
#   display(Image.fromarray(x_train[i]))
```

```
set1x = x_train[:1000]
set1y = y_train[:1000]
set2x = x_train[1000:2000]
set2y = y_train[1000:2000]
set3x = x_train[2000:3000]
set3y = y_train[2000:3000]
```

```
display(Image.fromarray(255 - set1x[0]))
```

```
# Batch 1
batch1x = []
batch1y = []
length = 4
boxwidth = 170 // length
move = boxwidth // 2 - num_w//2 - 5
for i in range(0, 1000, length):
  pos = (170 // length) // 2
  image = white.copy()
  label = ""
  for numimg,num in zip(set1x[i:i+length],set1y[i:i+length]):
    numimg = Image.fromarray(255 - numimg)
    numimg = numimg.resize((num_h, num_w), Image.ANTIALIAS)
    shiftw = random.randint(-move , move)
    shifth = random.randint(-2,2)
    image.paste(numimg, (pos + shiftw - num_w//2, 15 + shifth))
    pos += boxwidth
    label += str(num)
  batch1x.append(np.array(image).T)
  batch1y.append(label)
```

```
# Batch 2
batch2x = []
batch2y = []
length = 5
boxwidth = 170 // length
move = boxwidth // 2 - num_w//2 - 5
for i in range(0, 1000, length):
  pos = (170 // length) // 2
  image = white.copy()
  label = ""
  for numimg,num in zip(set2x[i:i+length],set2y[i:i+length]):
    numimg = Image.fromarray(255 - numimg)
    numimg = numimg.resize((num_h, num_w), Image.ANTIALIAS)
    
    shiftw = random.randint(-move , move)
    shifth = random.randint(-2,2)
    image.paste(numimg, (pos + shiftw - num_w//2, 15 + shifth))
    pos += boxwidth
    label += str(num)
  batch2x.append(np.array(image).T)
  batch2y.append(label)
```

```
# Batch 3
batch3x = []
batch3y = []
length = 6
boxwidth = 170 // length
move = boxwidth // 2 - num_w//2 - 5
for i in range(0, 1000, length):
  pos = (170 // length) // 2
  image = white.copy()
  label = ""
  for numimg,num in zip(set3x[i:i+length],set3y[i:i+length]):
    numimg = Image.fromarray(255 - numimg)
    numimg = numimg.resize((num_h, num_w), Image.ANTIALIAS)
    shiftw = random.randint(-move , move)
    shifth = random.randint(-2,2)
    image.paste(numimg, (pos + shiftw - num_w//2, 15 + shifth))
    pos += boxwidth
    label += str(num)
  batch3x.append(np.array(image).T)
  batch3y.append(label)
```

```
for image,label in zip(batch1x[:20], batch1y[:10]):
  display(ImageOps.expand(Image.fromarray(image.T), border=1, fill='black'))
  print(label)
for image,label in zip(batch2x[:10], batch2y[:10]):
  display(ImageOps.expand(Image.fromarray(image.T), border=1, fill='black'))
  print(label)
for image,label in zip(batch3x[:10], batch3y[:10]):
  display(ImageOps.expand(Image.fromarray(image.T), border=1, fill='black'))
  print(label)
```
